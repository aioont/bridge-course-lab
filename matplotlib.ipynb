```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
x=np.array([1,8])
y=np.array([3,5])
plt.plot(x,y,"*")
plt.show()
```

```python
x=np.array([1,3,6,8])
y=np.array([3,5,8,10])
plt.plot(x,y,"o")
plt.show()
```

```python
y=np.array([3,8,10,5,7])
plt.plot(y)
plt.show()
```

```python
# marker
x=np.array([1,3,5,7,9])
y=np.array([2,4,6,8,10])
plt.plot(x,y, marker="x",ms=20, mec="r")
plt.show()
```

```python
# line style
plt.plot(y,linestyle='dashed', color='r')
plt.show()

plt.plot(y,linestyle='dotted', linewidth='25')
plt.show()


plt.plot(y,linestyle='dashdot')
plt.show()
```

```python
z1=np.array([1,4,8,12,16])
z2=np.array([5,6,7,8,9])

plt.plot(z1)
plt.plot(z2)

plt.title("Sporti Data")

plt.xlabel("Average Calorie")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
x=np.array([1,3,5,7,9])
y=np.array([2,4,6,8,10])
plt.grid()
plt.subplot(1,2,1)
plt.plot(x,y)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y, width=0.5)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,height=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
marks=[90,5,4,44,1,4]
grade_interval=[0,35,70,100]
plt.title("Student Grade")
plt.hist(marks, grade_interval, histtype="bar", rwidth=0.7)
plt.xticks([0,35,70,100])
plt.xlabel("Percentage")
plt.ylabel("No. of students")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[10,50,100,80,90,60]
plt.xlabel("Percentage")
plt.ylabel("No. of students")
plt.scatter(x,y,c="r",s=200,alpha=0.5)
plt.show()
```
