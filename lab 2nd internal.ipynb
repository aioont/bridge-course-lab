```python
import numpy as np
arr=[]

n=int(input("Enter number of elements : "))
print("Enter elements : ")
for i in range(1,n+1):
    val=int(input())
    arr.append(val)
    
arr2=np.array([10,20,30,40])

s=int(input("Enter elements to search :"))
if s in arr:
    print("Element found")
else:
    print("Element not found")

join_array=np.concatenate((arr,arr2))
print("Concatenated array : ", join_array)

print("Sorted array : ", np.sort(arr))


arr2[1:6]=10
print("Broadcasted array = ", arr2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories=['Category A','Category B', 'Category C', 'Category D']
men_means=[40,26,35,37]
women_means=[25,32,30,35]
men_std=[4,3,4,1]
women_std=[3,5,2,3]

bar_width=0.35
x=np.arange(len(categories))

plt.bar(x, men_means, bar_width, yerr=men_std, label="Men", color='blue')
plt.bar(x, women_means, bar_width, yerr=women_std, label="Women", color='pink')

plt.xlabel('Categories')
plt.ylabel('Means')
plt.title('Stacked bar plot with error bars')
plt.xticks(x, categories)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 21]

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 30]

# Create subplots
fig, axes = plt.subplots(1, 2)

# Plot the line plot in the first subplot (axes[0])
axes[0].plot(x, y)
axes[0].set_xlabel('X-axis')
axes[0].set_ylabel('Y-axis')
axes[0].set_title('Line Plot')

# Plot the bar chart in the second subplot (axes[1])
axes[1].bar(categories, values)
axes[1].set_xlabel('Categories')
axes[1].set_ylabel('Values')
axes[1].set_title('Bar Chart')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
