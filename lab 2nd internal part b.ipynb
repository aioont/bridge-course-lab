```python
import matplotlib.pyplot as plt
lang=['Java','Python','C','REST','C#', 'C++', 'PHP']
values=[22,54,34,45,87,12,90]
def addLabels(values):
    for i in range(len(values)):
        plt.text(i, values[i], values[i])
        
plt.bar(lang, values, color=["red", "green", 'blue', 'violet', 'pink', 'yellow'], label=lang)
addLabels(values)
plt.legend()
plt.title("Summary")
plt.xlabel("Language")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=[10, 20, 30, 40, 20]
y=["Dead", "Killed", "Alive", "Not survived", 'Survived']
c=['red','blue','pink','black','gray']

plt.bar(y,x,color=c, label=y)

def addLabels(y):
    for i in range(len(y)):
        plt.text(i,x[i],x[i])
        

plt.xlabel("Time")
plt.ylabel("No of people")
plt.title("Datastictics")
plt.legend()
addLabels(y)

plt.show()
```
